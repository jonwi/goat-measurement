```python
from shapely import Polygon
from shapely.ops import unary_union
from shapely.plotting import plot_polygon
from shapely import get_exterior_ring
import numpy as np

def convert(shapes):
    polys = []
    for shape in shapes:
        arr = np.fromstring(shape, dtype=float, sep=" ")
        arr = arr[1:].reshape(-1, 2)
        poly = Polygon(arr)
        polys.append(poly)

    union = unary_union(polys)
    
    ring = get_exterior_ring(union)
    factor = 0.00005
    while ring == None:
        ring = get_exterior_ring(union.buffer(factor))
        factor *= 10
        
    xy = np.array(list(zip(*ring.coords.xy))).reshape(-1)
    string = np.array2string(xy).replace("\n", "").replace("[", "").replace("]", "")
    string = ' '.join(string.split())
    return string
```

```python
import os


def convert_directory(path):
    dirs = ["test", "valid", "train"]
    for dr in dirs:
        dirpath = os.path.join(path, dr, "labels")
        if os.path.isdir(dirpath):
            for item in os.listdir(dirpath):
                filepath = os.path.join(dirpath, item)
                if os.path.isfile(filepath):
                    convert_file(filepath)

    
def convert_file(filename):
    with open(filename) as f:
        content = f.readlines()

    try:
        conv = convert(content)
        with open(filename, "w") as f:
            f.write("0 " + conv)
            f.truncate()
    except Exception as e:
        print("cannot convert file ", filename)
        print(e)
        

```
